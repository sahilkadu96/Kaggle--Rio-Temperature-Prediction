```python

import numpy as np # linear algebra
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('station_rio.csv')
df
```

```python
df.iloc[0, 0], df.iloc[0, 1], 
```

```python
df.drop(['D-J-F', 'M-A-M', 'J-J-A', 'S-O-N', 'metANN'], axis = 1, inplace = True)
```

```python
temp  = []
for i in range(len(df)):
    for j in range(1, df.shape[1]):
        temp.append(df.iloc[i, j])
        
temp
    
```

```python
months = []
for i in range(len(df)):
    for j in range(1, 13):
        months.append(j)
        
months
        
```

```python
year  = []
for i in range(len(df)):
    for j in range(1, 13):
        year.append(df['YEAR'][i])
        
year
```

```python
df2 = pd.DataFrame({'Year': year, 'Month':months, 'Temperature':temp})
df2
```

```python
df2['Day'] = 1
df2.head()
```

```python
df2['Date'] = pd.to_datetime(df2[['Year', 'Month', 'Day']])
df2.head()
```

```python
df2.info()
```

```python
df2.drop(['Year', 'Month', 'Day'], axis = 1, inplace = True)
```

```python
df2.replace(999.9, np.nan, inplace = True)
```

```python
df3 = df2.set_index('Date', drop = True)
```

```python
df3.head()
```

```python
df4 = df3.interpolate()
```

```python
plt.figure(figsize = (20, 15))

plt.subplot(2,1,1)
plt.plot(df3, color = 'blue')
plt.title('Before interpolation')

plt.subplot(2,1,2)
plt.plot(df4, color = 'red', linestyle = '--')
plt.plot(df3, color = 'blue')
plt.title('After interpolation')
```

```python
df4.to_csv('station_rio_preprocessed.csv', index = False)
```
